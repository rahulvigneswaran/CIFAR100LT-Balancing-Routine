```python
import os

# CUDA devices used
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
```

```python
from cifar100lt import IMBALANCECIFAR100
import numpy as np
from utils import fine_labels
import PIL 

data_train = IMBALANCECIFAR100(phase="train", imbalance_ratio=0.01)

fl_map = np.array(fine_labels())
label, count = np.unique(data_train.labels, return_counts=True)

gen_count = [count[0] - i for i in count]

prompts = []
prompts_source_image_label = []
prompts_final_image_label = []
source_imgs = []
for i, j in zip(label, gen_count):
    for k in np.random.choice(np.arange(0,50), j, replace=True): #Choosing only from the top 50 head classes
        if i == k : 
            try:
                k+=1
            except:
                k-=1
        a = data_train.data[np.array(data_train.labels) == k]
        a_arr = a[np.random.randint(len(a))] #a[np.random.choice(np.arange(len(a)), 1)] np.random.randint(len(a))
        img = PIL.Image.fromarray(a_arr)
        source_imgs.append(img)
        prompts.append((fl_map[k], fl_map[i]))
        prompts_source_image_label.append(k)
        prompts_final_image_label.append(i)

prompts, prompts_source_image_label#, prompts_final_image_label, source_imgs
```

```python
seed=3624651140942376049
# torch.manual_seed(seed)
print('seed:',seed)
for i in range(len(prompts)):
    p = prompts[i]
    img = source_imgs[i]
    label = prompts_final_image_label[i] 
    print(p, label)
```
